```python
What is the name of the feature responsible for generating Regex objects? The re module in Python is responsible for generating Regex objects.

Why do raw strings often appear in Regex objects? Raw strings (e.g., r'...') are used in Regex objects to avoid escaping backslashes. This makes the pattern more readable.

python
pattern = r'\d\d\d'
What is the return value of the search() method? The search() method returns a match object if the pattern is found in the string; otherwise, it returns None.

From a Match item, how do you get the actual strings that match the pattern? You can use the group() method on the match object to get the matching string.

python
match = re.search(r'\d+', '123abc')
if match:
    print(match.group())  # Outputs: '123'
In the regex which created from the r'(\d\d\d)-(\d\d\d-\d\d\d\d)', what does group zero cover? Group 2? Group 1?

Group 0 covers the entire match.

Group 1 covers the first set of parentheses: \d\d\d

Group 2 covers the second set of parentheses: \d\d\d-\d\d\d\d

python
regex = re.compile(r'(\d\d\d)-(\d\d\d-\d\d\d\d)')
match = regex.search('123-456-7890')
if match:
    print(match.group(0))  # Outputs: '123-456-7890'
    print(match.group(1))  # Outputs: '123'
    print(match.group(2))  # Outputs: '456-7890'
In standard expression syntax, parentheses and intervals have distinct meanings. How can you tell a regex that you want it to fit real parentheses and periods? You can escape parentheses and periods with a backslash (\).

python
pattern = r'\(123\)\.456'
The findall() method returns a string list or a list of string tuples. What causes it to return one of the two options?

If the regex has no capturing groups, findall() returns a list of strings.

If the regex has capturing groups, findall() returns a list of tuples, where each tuple contains the strings matched by the capturing groups.

python
regex1 = re.compile(r'\d+')
print(regex1.findall('123 456 789'))  # Outputs: ['123', '456', '789']

regex2 = re.compile(r'(\d+)-(\d+)')
print(regex2.findall('123-456 789-012'))  # Outputs: [('123', '456'), ('789', '012')]
In standard expressions, what does the | character mean? The | character represents an OR operation, matching either the pattern on its left or right side.

python
pattern = re.compile(r'apple|banana')
In regular expressions, what does the . character stand for? The . character matches any single character except a newline.

In regular expressions, what is the difference between the + and * characters?

+ matches one or more occurrences of the preceding element.

* matches zero or more occurrences of the preceding element.

What is the difference between {4} and {4,5} in regular expressions?

{4} matches exactly four occurrences of the preceding element.

{4,5} matches between four and five occurrences of the preceding element.

What do the \d, \w, and \s shorthand character classes signify in regular expressions?

\d matches any digit (0-9).

\w matches any word character (alphanumeric and underscore).

\s matches any whitespace character (space, tab, newline).

What do \D, \W, and \S shorthand character classes signify in regular expressions?

\D matches any non-digit character.

\W matches any non-word character.

\S matches any non-whitespace character.

What is the difference between .*? and .*?

.*? is a non-greedy match, matching as few characters as possible.

.* is a greedy match, matching as many characters as possible.

What is the syntax for matching both numbers and lowercase letters with a character class? You can use a character class that includes both ranges.

python
pattern = re.compile(r'[a-z0-9]')
What is the procedure for making a normal expression in regex case insensitive? You can pass the re.IGNORECASE flag to the re.compile() function.

python
pattern = re.compile(r'pattern', re.IGNORECASE)
What does the . character normally match? What does it match if re.DOTALL is passed as 2nd argument in re.compile()?

Normally, the . character matches any single character except a newline.

With re.DOTALL, the . character matches any character, including a newline.

If numReg = re.compile(r'\d+'), what will numRegex.sub('X', '11 drummers, 10 pipers, five rings, 4 hen') return?

python
numRegex = re.compile(r'\d+')
result = numRegex.sub('X', '11 drummers, 10 pipers, five rings, 4 hen')
print(result)  # Outputs: 'X drummers, X pipers, five rings, X hen'
What does passing re.VERBOSE as the 2nd argument to re.compile() allow to do? It allows you to write more readable regex patterns by enabling you to include whitespace and comments.

python
pattern = re.compile(r'''
    \d+    # Match one or more digits
    -      # Followed by a hyphen
    \d+    # Match one or more digits
''', re.VERBOSE)
How would you write a regex that match a number with comma for every three digits?

python
pattern = re.compile(r'^\d{1,3}(,\d{3})*$')
This matches:

'42'

'1,234'

'6,368,745'

But not:

'12,34,567'

'1234'

How would you write a regex that matches the full name of someone whose last name is Watanabe?

python
pattern = re.compile(r'[A-Z][a-z]*\sWatanabe')
This matches:

'Haruto Watanabe'

'Alice Watanabe'

'RoboCop Watanabe'

But not:

'haruto Watanabe'

'Mr. Watanabe'

'Watanabe'

'Haruto watanabe'

How would you write a regex that matches a sentence where the first word is either Alice, Bob, or Carol; the second word is either eats, pets, or throws; the third word is apples, cats, or baseballs; and the sentence ends with a period?

python
pattern = re.compile(r'^(Alice|Bob|Carol) (eats|pets|throws) (apples|cats|baseballs)\.$', re.IGNORECASE)
This matches:

'Alice eats apples.'

'Bob pets cats.'

'Carol throws baseballs.'

'Alice throws Apples.'

'BOB EATS CATS.'

But not:

'RoboCop eats apples.'

'ALICE THROWS FOOTBALLS.'

'Carol eats 7 cats.'
```
